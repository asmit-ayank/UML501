```python
import numpy as np

def calculate_probability(x, mean, variance):
    epsilon = 1e-6
    variance += epsilon
    exponent = np.exp(-((x - mean)**2 / (2 * variance)))
    return (1 / (np.sqrt(2 * np.pi * variance))) * exponent

def NaiveBayes(X, Y):
    priors = {}
    means = {}
    variances = {}
    classes = np.unique(y_train)

    for c in classes:
        x_train_c = x_train[y_train == c]
        priors[c] = len(x_train_c) / len(x_train)
        means[c] = np.mean(x_train_c, axis=0)
        variances[c] = np.var(x_train_c, axis=0)
    y_pred = []
    for sample in x_test:
        posteriors = []
        for c in classes:
            log_likelihood = np.sum(np.log(calculate_probability(sample, means[c], variances[c])))
            posterior = np.log(priors[c]) + log_likelihood
            posteriors.append(posterior)
        predicted_class = classes[np.argmax(posteriors)]
        y_pred.append(predicted_class)

    return x_train, x_test, y_train, y_test, priors, means, variances, np.array(y_pred)
```

```python
x_train, x_test, y_train, y_test, priors, means, variances, y_pred_nb = NaiveBayes(X, Y)

accuracy_nb = accuracy_score(y_test, y_pred_nb)
print(f"Accuracy of Naive Bayes: {accuracy_nb:.4f}")
print(classification_report(y_test, y_pred_nb))
```
